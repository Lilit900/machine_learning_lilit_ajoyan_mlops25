### b) Draw a line chart of quantity for each year. Is there a year that should be skipped?



```python
quantity_per_year = sales.groupby('Year')['Quantity'].sum()
quantity_per_year= quantity_per_year.drop(2017)
quantity_per_year
```

```python
import matplotlib.pyplot as plt

quantity_per_year.plot(kind='line', marker='o')
plt.xlabel('Year')
plt.ylabel('Quantity')
plt.title('Quantity per Year')
plt.show()
```

###   c) Draw a line chart of average CO2 emissions for same years that as in b)



```python
avg_co2_emissions_per_year = sales.groupby('Year')['Avg_CO2'].sum()
avg_co2_emissions_per_year= avg_co2_emissions_per_year.drop(2017)
avg_co2_emissions_per_year
```

```python
avg_co2_emissions_per_year.plot(kind='line', marker='o')
plt.xlabel('Year')
plt.ylabel('Avg CO2 emission')
plt.title('Average CO2 emission per Year')
plt.show()
```

###  d) Draw a line chart of all years and months for import



```python
import_year_month = sales.groupby(['Year', 'Month'])['Import'].sum()
import_year_month
```

```python
import_year_month.plot(kind='line', marker='o')
plt.xlabel('Year and month')
plt.ylabel('import')
plt.title('Years_Months for import')
plt.show()
```

###   f) Draw a line chart of all years and months for electric cars import where it's relevant.

```python
electric_car_year_month = sales.groupby(['Year', 'Month'])['Import_Electric'].sum()
electric_car_year_month = electric_car_year_month[electric_car_year_month > 0]
electric_car_year_month
```

```python
electric_car_year_month.plot(kind='line', marker='o')
plt.xlabel('Year and month')
plt.ylabel('Import')
plt.title('Import of electrical cars')
plt.show()
```
