## Filter out a country and find its KPIs


```python
country = 'Canada'
df_by_country= df.query('Country.str.casefold()== @country.casefold()' )
df_by_country.head(2)
```

```python
str(df_by_country['Profit'].sum()), str(df_by_country['Cost'].sum()) , str(len(df_by_country))
```

make into dict

```python
{
    "total_profit": str(df_by_country["Profit"].sum()),
    "total_cost": str(df_by_country["Cost"].sum()),
    "number_of_purchases": str(len(df_by_country)),
}
```
